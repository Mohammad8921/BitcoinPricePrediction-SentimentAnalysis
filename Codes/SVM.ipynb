```python
import numpy as np
import pandas as pd
```

## Importing the Dataset 

```python
dataset = pd.read_csv('../dataset/dataset_sampled.csv')
```

## Creating Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000)
X = cv.fit_transform(dataset['cleaned']).toarray()
y = dataset['Labels']
```

## Creating The Training set and the Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

params = {'C':[1, 10], 'kernel':('poly','rbf','sigmoid')}

svc_classifier_grid = SVC()
svc_gridsearch = GridSearchCV(svc_classifier_grid, param_grid=params, scoring='accuracy', cv=10)
svc_gridsearch.fit(X_train,y_train)
```

```python
print("The best validation accuracy: ", svc_gridsearch.best_score_)
svc_gridsearch.best_estimator_.get_params()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
svc_classifier = svc_gridsearch.best_estimator_
y_pred_svm = svc_classifier.predict(X_test)
print(confusion_matrix(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```
